```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-1,1,50)
y1=x**2
y2=x*2+1

plt.figure()
plt.plot(x,y1)

plt.figure(num=3,figsize=(8,5))
plt.plot(x,y2)#默认宽度是1
plt.plot(x,y1,color='red',linewidth=5.0,linestyle='--')


plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel('i am x')
plt.ylabel('i am y')

new_ticks=np.linspace(-1,2,5)
plt.xticks(new_ticks)#更改x轴的刻度
plt.yticks([-2,-1,1,2,3],['so bad','bad','normal','good','so good'])
#plt.yticks([-2,-1,1,2,3],[r'$so\ bad$',r'$bad$',r'$normal$',r'$good$',r'$so\ good$'])
#转置符形式，改变字体，添加特殊符号

plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-1,1,50)
y1=x**2
y2=x*2+1

plt.figure(num=3,figsize=(8,5))
plt.plot(x,y2)#默认宽度是1
plt.plot(x,y1,color='red',linewidth=5.0,linestyle='--')


plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel('i am x')
plt.ylabel('i am y')

new_ticks=np.linspace(-1,2,5)
plt.xticks(new_ticks)#更改x轴的刻度
plt.yticks([-2,-1,1,2,3])


ax=plt.gca()#get current axis
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')#去掉上面和右边的框
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',-1))#把x轴移动到y轴-1的地方
ax.spines['left'].set_position(('data',0))



plt.show
```

```python
#lengend图例
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-1,1,50)
y1=x**2
y2=x*2+1



plt.figure()
plt.plot(x,y2)#默认宽度是1
plt.plot(x,y1,color='red',linewidth=5.0,linestyle='--')


plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel('i am x')
plt.ylabel('i am y')

new_ticks=np.linspace(-1,2,5)
plt.xticks(new_ticks)#更改x轴的刻度
plt.yticks([-2,-1,1,2,3])

#下面的逗号很重要
l1,=plt.plot(x,y2,label='up',color='blue')
l2,=plt.plot(x,y1,color='red',linewidth=5.0,linestyle='--',label='down')
plt.legend(handles=[l1,l2],labels=['a','b'],loc='best')         

plt.show
```

```python
#annotation 标注
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-3,3,50)
y=2*x+1

plt.figure(num=1,figsize=(8,5))
plt.plot(x,y)

ax=plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

x0=1
y0=2*x0+1
plt.plot([x0,x0],[0,y0],'k--',lw=2.5)
plt.scatter(x0,y0,s=50,color='b')#打印出x0y0那个点

#method1########
plt.annotate(r'$2x+1=%s$'%y0,xy=(x0,y0),xycoords='data',
            xytext=(+30,-30),textcoords='offset points',fontsize=16,
            arrowprops=dict(arrowstyle='->',connectionstyle="arc3,rad=.2"))
#method2########
plt.text(-3.5,3,r'$This\ is\ the\ some\ text. \mu\ \sigma_1\ \alpha_2$',
        fontdict={'size':16,'color':'r'})
plt.show()
            
```

```python
#axis ticks 让刻度可见
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-3,3,50)
y=0.1*x

plt.figure
plt.plot(x,y,lw=10,zorder=1)
plt.ylim(-1,1)
ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

for label in ax.get_xticklabels()+ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white',edgecolor='red',alpha=0.8,zorder=2))
    #alplha透明度

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

n=1024
X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)
T=np.arctan2(Y,X)

plt.scatter(X,Y,s=75,c=T,alpha=0.5)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.xticks(())
plt.yticks(())#去掉xy的刻度

plt.show()

```

```python
#bar 柱状图
import matplotlib.pyplot as plt
import numpy as np

n=12
X=np.arange(n)
Y1=(1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2=(1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.2,y+0.01,'%.2f'%y,ha='center',va='bottom')
    #ha水平对齐，va竖直对齐(向下对齐)

for x,y in zip(X,Y2):
    plt.text(x+0.2,-y-0.01,'-%.2f'%y,ha='center',va='top')
    
plt.show()
```

```python
#contours 等高线轮廓
import matplotlib.pyplot as plt
import numpy as np

def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n=256
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)
X,Y=np.meshgrid(x,y) #产生一个网格底面

plt.contourf(X,Y,f(X,Y),8,alpha=.75,cmap=plt.cm.cool)
# #fill contour ,填充轮廓，8是等高线的数量

C=plt.contour(X,Y,f(X,Y),8,colors='black',lw=.5)
# #add contour lines
plt.clabel(C,inline=True,fontsize=10)

plt.xticks(())
plt.yticks(())
plt.show()
```
